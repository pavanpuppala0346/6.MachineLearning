```python
import pandas as pd

df = pd.read_csv("iris.csv")
df.head()
```

```python
df.shape
```

#### univariate Analysis

```python
        #object[condition]
df_setosa = df.loc[df['species']=='setosa']
df_virginica = df.loc[df['species']=='virginica']
df_versicolor= df.loc[df['species']=='versicolor']
```

```python
import numpy as np

df_setosa['sepal_length']
np.zeros_like(df_setosa['sepal_length'])
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.plot(df_setosa['sepal_length'], np.zeros_like(df_setosa['sepal_length']),'o')
plt.plot(df_virginica['sepal_length'], np.zeros_like(df_virginica['sepal_length']),'o')
plt.plot(df_versicolor['sepal_length'], np.zeros_like(df_versicolor['sepal_length']),'o')
plt.xlabel('petal_length')
plt.show()
```

#### bivariate analysis

```python
import seaborn as sns

sns.FacetGrid(df, hue='species', height=4).map(plt.scatter,'sepal_length','sepal_width')
plt.show()
```

```python
import seaborn as sns

sns.pairplot(df, hue='species', height=4)
plt.show()
```
