#### Step2: calculating Euclidean distance between two vectors

```python
import numpy as np

def ed(known_row,unknown_row):
    distance=0.0
    for i in range(len(known_row)-1):
        distance = distance+(known_row[i]-unknown_row[i])**2
    return np.sqrt(distance)
```

```python
X = df.values
```

```python
print(type(X))
print(X.ndim)
print(X.shape)
```

```python
unknown_row = X[1]
unknown_row
```

```python
for known_row in X:
    distance = ed(known_row,unknown_row)
    print(distance)
```

#### step3: Locate the most nearest neighbors

```python
def get_neighbors(known_row, unknown_row, n_neighbors):
    distances = []                             # empty list
    for row in known_row:                      # from 150 samples train data
        dist = ed(row, unknown_row)
        #print(row)
        #print(unknown_row)
        distances.append((row, dist))
    #print(distances)
    distances.sort(key=lambda tup:tup[1])
    neighbors=[]
    for i in range(n_neighbors):
        neighbors.append(distances[i][0])
    return neighbors
```

#### step4: making a classification prediction with neighbors

```python
unknown_row = [[5.1, 3.5, 1.4, 0.2],
              [5.1, 3.5, 1.4, 0.3],
              [5.0, 3.6, 1.4, 0.2]]
```

```python
def predict_classification(known_row, unknown_row, n_neighbors):
    rs=[]
    for uks in unknown_row:
        neighbors = get_neighbors(known_row, uks, n_neighbors)
        #print(neighbors)
        output_values = [row[-1] for row in neighbors]
        prediction = max(set(output_values), key=output_values.count)          # max species
        rs.append(prediction)
    return rs
```

```python
predict_classification(X, unknown_row, 3)
```

```python
pred = predict_classification(X, unknown_row, 3)
print(f"real value: {X[0][-1]}, Model predicted value: {pred}")
```
